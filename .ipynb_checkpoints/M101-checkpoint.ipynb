# Redenning laws
---
0: Calzetti  
1: Leitherer  
2: Calzetti extrapol from Reddy+16  
3: Reddy+16   
4: SMC

---


```python
def ElamV_FM(wa, c1, c2, c3, c4, x0, gamma):
    """ Base function for Extinction curves that use the form from
    Fitzpatrick & Massa 90.

    Parameters
    ----------
    wa : array_like
      One or more wavelengths in Angstroms.
    c1, c2, c3, c4, gamma, x0 : float
      Constants that define the extinction law.

    Returns
    -------
    ElamV : ndarray or float if `wa` is scalar
       extinction in the form E(lambda - V) / E(B - V)
    """

    #x = 1e4 / np.array(wa, ndmin=1)
    xm = 1./wa
    xsq = xm*xm
    D = xsq / ((xsq - x0*x0)**2 + xsq*gamma*gamma)
    FMf = c1 + c2*xm + c3*D

    if len(xm) == 1:
        if xm >= 5.9:
            FMf += c4*(0.5392*(xm - 5.9)**2 + 0.05644*(xm - 5.9)**3)
        ElamV = FMf[0]
    else:
        c4m = xm >= 5.9
        FMf[c4m] += c4*(0.5392*(xm[c4m] - 5.9)**2 + 0.05644*(xm[c4m] - 5.9)**3)
        ElamV = FMf

    return ElamV

    
def dustAtt(wave, ebv, choice):
    xmicr= wave/1e4
    #wave /= 1e4
    if choice == 0: #calz
        params = [2.659,-2.156,1.509,-0.198,0.011,4.05]
        kl  = params[0]*(params[1]+(params[2]/xmicr)+(params[3]/(xmicr**2.))+(params[4]/(xmicr**3.)))+params[5]
    elif choice == 1: #leitherer
        params = [5.472, 0.671, -9.218e-3, 2.62e-3]
        kl  = params[0]+(params[1]/xmicr)+(params[2]/(xmicr**2.))+(params[3]/(xmicr**3.))
    elif choice == 2: #calzetti from reddy's extrapolation
        params = [4.126, 0.931]
        kl = params[0]+(params[1]/xmicr)
    elif choice == 3: #Reddy
        params = [2.191, 0.974]
        kl = params[0]+(params[1]/xmicr)
    elif choice == 4: #SMC
        c1,c2,c3,c4 = -4.959, 2.264, 0.389, 0.461
        x0, gamma = 4.6, 1.0
        ElamV = ElamV_FM(xmicr, c1, c2, c3, c4, x0, gamma)
        AlamAv = (ElamV + 2.74) / 2.74 
        kl = AlamAv * 2.74
        
    return 10**(-0.4*kl*ebv)


def ccm_unred(wave, flux, ebv):
    R_V = 3.1

    x = 10000./ wave           
    npts = len(x)
  #  a = np.zeros(npts)  
 #   b = np.zeros(npts)

    c1 = [ 1. , 0.104,   -0.609,    0.701,  1.137,   
             -1.718,   -0.827,    1.647, -0.505 ]        
    c2 = [ 0.,  1.952,    2.908,   -3.989, -7.985,   
             11.102,    5.491,  -10.805,  3.347 ]
    y = x - 1.82
    p1 = np.poly1d(np.flip(c1))
    p2 = np.poly1d(np.flip(c2))

    a = p1(y)
    b = p2(y)
 

    A_V = R_V * ebv
    A_lambda = A_V * (a + b/R_V)
 
    funred = flux * 10.**(0.4*A_lambda)   

    return funred
    
```

## Read spectra

```python
# Load data from file
file_path = 'CONT_FITS_JPC/REG2-sb99-fit.txt'
data = np.genfromtxt(file_path, delimiter=None)  # Ensure correct delimiter

# Extract columns (wavelength, flux, error, fitted values)
wavelengths = data[:, 0]
flux = data[:, 1]
error = data[:, 2]
fitted_flux = data[:, 3][wavelengths>500]

# Normalize flux using the median in the range 1250-1300
norm_value = np.nanmedian(flux[(wavelengths > 1250) & (wavelengths < 1300)])

if norm_value == 0 or np.isnan(norm_value):
    raise ValueError("Normalization value is zero or NaN, check input data!")

flux /= norm_value
error /= norm_value
fitted_flux /= norm_value

# Plot spectrum
plt.figure(figsize=(10, 2))
plt.plot(wavelengths, flux, label="Observed Flux", color="black")
plt.xlim(1500, 1600)
plt.ylim(0, 2)
plt.xlabel("Wavelength ()")
plt.ylabel("Normalized Flux")
plt.title("Normalized Spectrum")
plt.legend()
plt.grid(True)

# Observational parameters
wave_obs = wavelengths[wavelengths>500]
flux_obs = flux[wavelengths>500]
err_obs = error[wavelengths>500]
FWHM_obs = 375  # Full Width at Half Maximum in km/s

# Calculate observational resolution
R_mod = 4000.0
sigma_obs = FWHM_obs / 2.3548  # Convert FWHM to standard deviation
R_obs = 299792.458 / FWHM_obs  # Observed spectral resolution

# Display the plot
plt.show()
```

## Loading mods

```python
def create_model_dict( data, name, R=1000):
    """
    Create a dictionary containing SSP model data.

    Parameters:
    - data (astropy.io.fits.fitsrec.FITS_rec): Data from the FITS file.
    - name (str): Name of the model.
    - R (int): Resolution parameter. Default is 1000.

    Returns:
    - model_dict (dict): Dictionary containing model data.
    """
    model_dict = {
        'wave': data['WAVE'][0],
        'flux': data['FLUX'][0],
        'age': np.round(data['AGE'][0], 2),
        'metallicity': np.round(data['ZSTELLAR'][0], 3),
        'R': R
    }
    return model_dict
```

```python
print("...Loading default SSP models (SB99 and BPASS)...")

with fits.open('./../SpectroFan/spectroFan/SSP_models/SB99_BPASS_50.fits') as hdu:
    sb99_data = hdu[1].data
    bpass_data = hdu[2].data

sb99_models = create_model_dict(sb99_data, 'sb99', R=3750)
bpass_models = create_model_dict(bpass_data, 'bpass', R=1500)

print("...Default SSP models loaded., Interp to good R")


diff_sigma = np.sqrt((1500/R_obs)**2 - (1500/R_mod)**2) / 2.354820046
pix_size_model_sb99 = np.abs(sb99_models['wave'][1] - sb99_models['wave'][0])
pix_size_model_bpass = np.abs(bpass_models['wave'][1] - bpass_models['wave'][0])

sigma_gaussian_sb99 = diff_sigma / pix_size_model_sb99
sigma_gaussian_bpass = diff_sigma / pix_size_model_bpass

sb99_mods_obs = np.zeros((len(sb99_models['flux']), len(wave_obs)))
for i in range(len(sb99_models['flux'])):
    sb99_mods_obs[i] = np.interp(
        wave_obs,
        sb99_models['wave'],
        gaussian_filter(sb99_models['flux'][i], sigma_gaussian_sb99)
    )

bpass_mods_obs = np.zeros((len(bpass_models['flux']), len(wave_obs)))
for i in range(len(bpass_models['flux'])):
    bpass_mods_obs[i] = np.interp(
        wave_obs,
        bpass_models['wave'],
        gaussian_filter(bpass_models['flux'][i], sigma_gaussian_bpass)
    )

plt.figure(figsize = (15, 4))
plt.plot(bpass_models['wave'], bpass_models['flux'][0])
plt.plot(wave_obs, bpass_mods_obs[0])
plt.xlim(1000, 2000)
```

## FITTING SB 99

```python
def model_lmfit(params,x,y, yerr, mods, mask):
    model = np.zeros(len(x))
    for i in range(len(mods)-10): # -10 to exclude 2Zsolar models
        model += params["a"+str(i)]*mods[i][mask]
    att = dustAtt(x, params["ebv"], 3) #3 for Reddy attenuation law
    model *= att
    return (y-model)/yerr
```

### MonteCarl Error

```python
#Monte Carlo fits 100 times
n_it = 100
ebv_arr = np.zeros(n_it)
age_arr = np.zeros(n_it)
metal_arr = np.zeros(n_it)

for ii in range(n_it):
    params = Parameters()
    for i in range(len(sb99_mods_obs)):
        params.add("a"+str(i), value =np.random.uniform(0,10), min =0.0, max=1000.0)  

    params.add("ebv", value = 0.2, min =0.0, max = 0.5)

    # Calling the minimize function. Args contains the x and y data. 
    mod_flux = ydataStell + yerrStell*np.random.normal(size= len(yerrStell))
    #Other methods to try: method='' with lssquare, powell, or bashinhopping
    fitted_params = minimize(model_lmfit, params, args=(xdataStell,mod_flux,yerrStell, sb99_mods_obs, good),
                        nan_policy='omit',reduce_fcn='neglogcauchy' )
    curcoefs=np.zeros(len(sb99_mods_obs)-10)
    for i in range(len(sb99_mods_obs)-10):
        curcoefs[i] = fitted_params.params["a"+str(i)].value
    ebv_arr[ii] = fitted_params.params["ebv"].value
    for i in range(len(sb99_models['metallicity'])-1):##CHANGED
        dataHist[i] = curcoefs[10*i:10*(i+1)]/sum(curcoefs)
        cumdataHist +=dataHist[i]
    print('Stellar weighted age %f Myrs'%(np.sum( np.multiply(bins,dataHist))))
    print('Stellar weighted metallicity %f'%(np.sum( np.multiply(metall_50.flatten(),curcoefs))/np.sum(curcoefs)))
    age_arr[ii] = np.sum( np.multiply(bins,dataHist))
    metal_arr[ii] =np.sum( np.multiply(metall_50.flatten(),curcoefs))/np.sum(curcoefs)
    print('Fit %d done:'%ii)
    
print('MonteCarlo Median Stellar weighted age %.2f +- %.2f  Myrs'%(np.median(age_arr), np.std(age_arr)))
print('MonteCarlo Median Stellar weighted metallicity %.3f +- %.3f '%(np.median(metal_arr), np.std(metal_arr)))
print('MonteCarlo Median E(B-V) %f +-  '%(np.median(ebv_arr)), np.std(ebv_arr))

ebv_sb99_err = np.std(ebv_arr)
age_sb99_err = np.std(age_arr)
met_sb99_err =  np.std(metal_arr)
```

```python
def model_lmfit_single(params,x,y, yerr, mods, mask):
    model = np.zeros(len(x))
    model += params["a0"]*mods[mask]
    att = dustAtt(x, params["ebv"], 3)
    model *= att
    return (y-model)/yerr
```

```python
best_chi = 1e10
best_idx = -1
best_params = None

for i in range(40):
    params = Parameters()
    params.add("a0", value =0.01, min =0.0, max=1e5)
    params.add("ebv", value = 0.1, min =00.0, max = 0.8)
    fitted_params = minimize(model_lmfit_single, params,
                             args=(xdataStell,ydataStell,yerrStell, sb99_mods_obs[i], good),
                        nan_policy='omit' )
    if(fitted_params.redchi < best_chi):
        best_chi = fitted_params.redchi
        best_idx = i
        best_params = fitted_params

```

```python
def model_lmfit(params,x,y, yerr, mods, mask):
    model = np.zeros(len(x))
    for i in range(len(mods)-10):
        model += params["a"+str(i)]*mods[i][mask]
    att = dustAtt(x, params["ebv"], 3)
    model *= att
    return (y-model)/yerr
```
